```python
skin_type_map = {
    "민감성": "N",
    "지성": "O",
    "건성": "A",
    "없음" : "NONE"
}
```

```python
## 전처리 안된파일 다시하기 
#new_df = pd.read_csv("./")
```

```python
# # 3. ingredeint INSERT (id는 auto_increment, 시간은 NOW())
for _, row in df.iterrows():
    # ingredient 테이블 insert
    insert_ingredient_sql = """
        INSERT INTO ingredient (created_at, updated_at, effect, name, risk, risk_category, ranking)
        VALUES (NOW(), NOW(), %s, %s, %s, %s, %s)
    """
    ingredient_values = (row["effect"], row["name"], row["risk"], row["risk_category"], row["ranking"])
    cursor.execute(insert_ingredient_sql, ingredient_values)

    # 방금 insert된 ingredient id
    ingredient_id = cursor.lastrowid

    # skin_type 한글 -> 약어로 매핑
    skin_types_kor = row["skin_type"].split(",")  # 예: "민감성,지성"
    for skin_kor in skin_types_kor:
        skin_kor = skin_kor.strip()
        skin_abbr = skin_type_map.get(skin_kor)
        if skin_abbr:  # 매핑 성공한 경우만 insert
            insert_caution_sql = """
                INSERT INTO caution_skin_type (ingredient_id, skin_type, created_at, updated_at)
                VALUES (%s, %s, NOW(), NOW())
            """
            cursor.execute(insert_caution_sql, (ingredient_id, skin_abbr))
        else:
            print(f"⚠️ 알 수 없는 피부타입: '{skin_kor}'")


# # 4. caution_skin_type에 insert
conn.commit()
print(f" {cursor.rowcount}개 레코드 삽입 완료")
cursor.close()
```
